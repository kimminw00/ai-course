```python
model = tf.keras.models.Sequential()
```

```python
vocab_size = 10000

model.add(tf.keras.layers.Embedding(vocab_size, 16, input_shape=(None,))) 
# input size가 다양해서 None을 사용함

model.add(tf.keras.layers.GlobalAveragePooling1D())
model.add(tf.keras.layers.Dense(16, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(1, activation=tf.nn.sigmoid))
# regression일때는 activation 안씀
# sigmoid 쓰니까 1로도 구분 가능 : 노드 하나만 있어도 2개 구분 가능
# softmax쓸라면 2로 해야한다.

# 왜 None쓰는거지?
# padding을 시켜도 사이즈를 예측할 수 없어서 None씀, (None,)을 씀
```

```python
model.compile(optimizer='Adam',
              # 인자 넣어서 초기값 받을 수 있다.
              # loss함수를 어떻게 최적화 시킬 지 하는 것
              loss='binary_crossentropy',  # 기본값을 넣어줌
              metrics=['acc'])
```
