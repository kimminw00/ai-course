```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential()
```

```python
iris.shape
```

```python
def create_model(op='adam'): # default값을 줘서 인자가 없어도 kc = KerasClassifier(create_model)를 넣을 수 있다.
    # def create_model(op='adam'): op가 parameter니까 param찾기 위해서 GridSearchCV 사용가능
    model = tf.keras.models.Sequential()
    model.add(tf.keras.layers.Dense(16, activation=tf.nn.relu, input_shape=(4,)))
    model.add(tf.keras.layers.Dense(3, activation=tf.nn.softmax)) 
    # 값이3이기 때문에, one-hot encoding해줘야 한다. (무슨 뜻이지?)
    model.compile(optimizer=op, loss='categorical_crossentropy', metrics=['acc'])
    return model
```

```python
import pandas as pd
```

```python
y = pd.get_dummies(iris.species)
```

```python
X = iris.iloc[:,:-1]
```

```python
model.fit(X, y, epochs=5)
# 왜 에러나지..? keras, tensorflow는 직접적으로 pandas 못 씀
# tf.data.dataset 써야 pandas keras써야 연동됨
# pandas를 numpy로 바꿔야 사용가능
```

```python
model.fit(X.values, y.values, epochs=5)
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
```

```python
kc = KerasClassifier(create_model)
```

```python
kc.fit(iris.iloc[:,:-1], y, epochs=20)
# pandas를 오류없이 사용할 수 있다.
# scikit에서 제공하는거 사용 가능 == cross validation, grid-search cv
# nueral network에서 pipeline, grid-search-cv가능
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(kc, X, y, cv=10)
# 연동 참 잘된다...!
```

```python
from sklearn.model_selection import learning_curve
```

```python
train_size, train_score, test_score = learning_curve(kc, X, y, cv=10)
```

```python
from sklearn_evaluation import plot
```

```python
plot.learning_curve(train_score, test_score, train_size)
# 그림도 잘 그려짐...! 내가 나중에 완성해야겠다.
```
