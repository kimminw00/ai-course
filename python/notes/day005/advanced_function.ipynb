# first class 함수면 반드시 higher order function이다.
# 함수의 인자로 함수를 넘겨줄 수 있음
# 함수의 return값으로 함수를 사용할 수 있음
함수에 함수를 넣으면, 원래 함수의 기능을 추가할 수 있음
ex) map 함수

```python
def greet(func):
    greeting = func('Hi, I am a Python Programmer')
    print('문근영')
    print(greeting)
```

```python
greet(1)
# int is not callable == 괄호를 사용할 수 없다.
```

```python
greet(bark)
```

```python
def whisper(text):
    return text.lower() + '...'
```

```python
greet(whisper)
```

```python
def speak(text):
    def whisper(t):
        return t.lower() + '...'
    return whisper(text)
# 중첩함수면 return 값을 본다. 함숫값인지 함수인지 확인한다.
```

```python
speak('Hello')
```

```python
whisper('Yo')
# speak 내부에 정의되어 있어서 whisper은 local함수이다.
# 함수 밖에서 함수 내부의 것을 접근할 수 없다. encapsulation
```

```python
speak.whisper
```

```python
__main__.speak.<locals>.whisper
```

```python
def get_speak_func(volume):
    def whisper(text):
        return text.lower()+'...'
    def yell(text):
        return text.upper()+'!..'
    if volume > 0.5:
        return yell
    else:
        return whisper
    
# 중첩함수에서는 함수를 return하는지 값을 return하는지 확인하기
```

```python
get_speak_func(0.3)('1')
```

```python
get_speak_func(0.7)
```

```python
speak_func = get_speak_func(0.7)
```

```python
speak_func('Hello')
```
