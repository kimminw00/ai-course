```python
import statsmodels.api as sm  # api는 R관례상 붙음
```

```python
dir(sm)
```

```python
sm.datasets.cancer.DESCRLONG
```

```python
dat = sm.datasets.get_rdataset("Guerry", 'HistData').data # .data하면 pandas format이 된다.
# cache기법으로 저장 가능, 매 번 인터넷에서 안 불러와도 됨
```

```python
dat
```

scikit에는 시계열 분석이 없음  
따라서 statsmodels & facebook prophet사용함

```python
import statsmodels.formula.api as smf  # formula == Formula-compatible models이 있는 패키지
results = smf.ols('Lottery ~ Literacy', data=dat).fit()  # scikit과 다른 방식으로 fit함
# 'Lottery ~ Literacy' : 는 dataframe column

# Lottery가 종속
# Literacy랑 Lottery가 어떤 관계인가?
```

```python
results.summary()
# 설명 용도로 사용
```

== -0.5245 * Literacy + 64.0896 == Lottery라는 관계가 있다.
intercept == y절편
coef가 기울기
